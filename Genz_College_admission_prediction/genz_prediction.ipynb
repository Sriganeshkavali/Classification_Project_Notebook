###  SEPARATING FEATURES AND TARGET

```python
X = df.drop(columns=['student_id', 'admission_status'])
y = df['admission_status']
```

### STRATIFIED SAMPLING FOR SPEED (Optional but Recommended)

```python
from sklearn.model_selection import train_test_split
```

```python
_, X_sample, _, y_sample = train_test_split(X, y, test_size=0.10, stratify=y, random_state=42)
print(f"Sample size for modeling: {X_sample.shape}")
```

### IDENTIFYING COLUMN TYPES

```python
numeric_features = X.select_dtypes(include=['int64', 'float64']).columns.tolist()
categorical_features = X.select_dtypes(include=['object']).columns.tolist()

print(f"Numerical Features: {numeric_features}")
print(f"Categorical Features: {categorical_features}")
```

### CREATING PREPROCESSING PIPELINES

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
#Standard scaler for numbers
numeric_transformer = Pipeline(steps=[
    ('scaler', StandardScaler())
])

categorical_transformer = Pipeline(steps=[
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
```

```python
# One-Hot Encoding for categorical text (like 'state' or 'gender')
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),
        ('cat', categorical_transformer, categorical_features)
    ])
```

### TRAIN-TEST SPLIT

```python
X_train, X_test, y_train, y_test = train_test_split(X_sample, y_sample, test_size=0.2, stratify=y_sample, random_state=42)
```

### Model Training and Selection
* We will test three distinct algorithms that represent different mathematical approaches:

1. Logistic Regression: A fast, interpretable, linear baseline model.

2. Random Forest: A robust, non-linear ensemble tree model.

3. LightGBM (Gradient Boosting): A state-of-the-art tree boosting algorithm designed specifically to be ultra-fast and high-performing on large tabular datasets.

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, roc_auc_score, confusion_matrix
import lightgbm as lgb

from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
```

```python
# DEFINE CANDIDATE MODELS

models = {
    "Logistic Regression": LogisticRegression(max_iter=1000, random_state=42),
    "Random Forest": RandomForestClassifier(n_estimators=100, max_depth=12, random_state=42, n_jobs=-1),
    "LightGBM": lgb.LGBMClassifier(n_estimators=100, learning_rate=0.05, random_state=42, verbose=-1)
}
```

```python

results = {}

## LOOP, TRAIN, AND EVALUATE
for name, model in models.items():
    print(f"\n Training {name}...")
    
    # Bundle preprocessing and modeling into a clean pipeline
    clf_pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                                  ('classifier', model)])
    
    # Train the model
    clf_pipeline.fit(X_train, y_train)
    
    # Predict probabilities and hard classes
    y_pred = clf_pipeline.predict(X_test)
    y_prob = clf_pipeline.predict_proba(X_test)[:, 1]
    
    # Calculate performance metrics
    accuracy = accuracy_score(y_test, y_pred)
    auc_roc = roc_auc_score(y_test, y_prob)
    
    results[name] = {
        "Pipeline": clf_pipeline,
        "Accuracy": accuracy,
        "AUC-ROC": auc_roc,
        "Predictions": y_pred,
        "Probabilities": y_prob
    }
    
    print(f" {name} Performance -> Accuracy: {accuracy:.4f} | AUC-ROC: {auc_roc:.4f}")
```

### COMPARE METRICS IN A TABLE

```python
df_perf = pd.DataFrame(results).T[['Accuracy', 'AUC-ROC']].astype(float)
print("\n--- Model Performance Comparison Table ---")
display(df_perf.sort_values(by="AUC-ROC", ascending=False))
```

```python
# Identify the winner automatically
best_model_name = df_perf['AUC-ROC'].idxmax()
print(f"\n🏆 The Best Model is: {best_model_name} (Based on AUC-ROC score)")
```

### PLOT CONFUSION MATRICES

```python

fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for idx, (name, res) in enumerate(results.items()):
    cm = confusion_matrix(y_test, res["Predictions"])
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', ax=axes[idx], cbar=False)
    axes[idx].set_title(f'{name}\nConfusion Matrix')
    axes[idx].set_xlabel('Predicted labels')
    axes[idx].set_ylabel('True labels')

plt.tight_layout()
plt.show()
```

```python
print(f"\n--- Detailed Classification Report for the Champion ({best_model_name}) ---")
print(classification_report(y_test, results[best_model_name]["Predictions"]))
```

```python
# Extract the winning pipeline
winner_pipeline = results[best_model_name]["Pipeline"]

# Get feature names after one-hot encoding transformation
cat_encoder = winner_pipeline.named_steps['preprocessor'].named_transformers_['cat'].named_steps['onehot']
encoded_cat_features = cat_encoder.get_feature_names_out(categorical_features).tolist()
all_features = numeric_features + encoded_cat_features

# Check if the winning model has feature importances feature
if hasattr(winner_pipeline.named_steps['classifier'], 'feature_importances_'):
    importances = winner_pipeline.named_steps['classifier'].feature_importances_
    
    # Create Dataframe
    feat_imp_df = pd.DataFrame({'Feature': all_features, 'Importance': importances})
    feat_imp_df = feat_imp_df.sort_values(by='Importance', ascending=False).head(10) # Top 10
    
    # Plot
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Importance', y='Feature', data=feat_imp_df, palette='rocket')
    plt.title(f'Top 10 Crucial Features Predicting College Admission ({best_model_name})')
    plt.xlabel('Relative Importance Value')
    plt.ylabel('Feature Name')
    plt.show()
else:
    # Fallback if Logistic Regression won (using coefficients instead)
    coefs = winner_pipeline.named_steps['classifier'].coef_[0]
    feat_imp_df = pd.DataFrame({'Feature': all_features, 'Coefficient': coefs})
    feat_imp_df['Abs_Coef'] = feat_imp_df['Coefficient'].abs()
    feat_imp_df = feat_imp_df.sort_values(by='Abs_Coef', ascending=False).head(10)
    
    plt.figure(figsize=(10, 6))
    sns.barplot(x='Abs_Coef', y='Feature', hue='Feature',data=feat_imp_df, palette='mako',legend=False)
    plt.title('Top 10 Crucial Features Predicting College Admission (LogReg Absolute Coefs)')
    plt.xlabel('Absolute Coefficient Weight')
    plt.ylabel('Feature Name')
    plt.show()
```
